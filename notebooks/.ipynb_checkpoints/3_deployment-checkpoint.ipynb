```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
import spacy
import pickle
from wordcloud import WordCloud
```

## Prediction

```python
# Load TF-IDF, chi-squared selector and model pkl for predict
model = pickle.load(open("../models/logistic_regression_model.pkl", "rb"))  
vectorizer = pickle.load(open("../models/feature_tf-idf.pkl", "rb"))  
chi2_selector = pickle.load(open("../models/chi2_selector.pkl", "rb"))  
```

```python
# Function for preprocessing
nlp = spacy.load("en_core_web_sm", disable=["parser", "ner"])

# Add other stopword
custom_stopwords = {"feel", "thing", "think", "know", "people", "make", "want", "need", "get", "go", "time", 
                    "feeling", "feelin", "feels", "felt", "feelings", "thinking"}

def text_preprocessing(text):
    # Remove URL, number, punctuation, and common contractions
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    text = re.sub(r'[-+]?[0-9]+', '', text)
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\b(don|doesn|didn|can|couldn|shouldn|won|wouldn|isn|aren|wasn|weren|hasn|haven|hadn|mustn|needn|oughtn|shan|mightn|mayn)\s+t\b', 
                  lambda match: f"{match.group(1)}t", text, flags=re.IGNORECASE)
    text = re.sub(r"\b(?:im|ive|ill|is|youre|theyre|whats|'s|s|'m|m|'ve|ve|'ll|ll)\b", '', text, flags=re.IGNORECASE)
    
    # Text casefolding, stopword removal, and lemmatization
    doc = nlp(text.lower())
    tokens = [token.lemma_ for token in doc if not token.is_stop and token.is_alpha and token.lemma_ not in custom_stopwords]
    tokens = [token for token in tokens if token != 's']
    
    return " ".join(tokens)

# Decode int label category to string 
emotion_mapping = {
    0: "Sadness",
    1: "Joy",
    2: "Love",
    3: "Anger",
    4: "Fear",
    5: "Surprise"
}
```

```python
def predict_emotion(input_text):
    # Preprocess input text
    clean_text = text_preprocessing(input_text)
    
    # Transform TF-IDF and Chi-Square features
    tfidf_features = vectorizer.transform([clean_text])
    selected_features = chi2_selector.transform(tfidf_features)
    
    # Perform prediction
    prediction = model.predict(selected_features)
    confidence = np.max(model.predict_proba(selected_features))
    predicted_emotion = emotion_mapping[prediction[0]]
    
    return predicted_emotion, confidence
```

```python
# User input
user_input = input("Enter your text: ")

# Predict and display result
predicted_emotion, confidence = predict_emotion(user_input)
print(f"Predicted Emotion: {predicted_emotion}")
print(f"Confidence: {confidence:.2f}")
```

## WordCloud

```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt

data=pd.read_csv('../data/emotions_clean.csv')
data = data.dropna()

data['label_cat'] = data['label'].map(emotion_mapping)
emotion_texts = data.groupby('label_cat')['clean_text'].apply(lambda x: ' '.join(x)).to_dict()
bg_color = "white"

for label, text in emotion_texts.items():
    # Buat WordCloud
    wordcloud = WordCloud(width=800, height=400, background_color=bg_color, colormap='viridis').generate(text)
    
    # Plot WordCloud
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.title(f"WordCloud for {label}", fontsize=16)
    plt.show()
```

This project focuses on developing an advanced emotion detection system using machine learning, trained on a large-scale dataset of 416,809 text samples labeled across six emotion categories: sadness, joy, love, anger, fear, and surprise.
